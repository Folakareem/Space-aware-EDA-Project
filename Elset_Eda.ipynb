Here are explanations for the different orbit classifications:

**Low Earth Orbit (LEO):**
*   **Altitude:** Typically between 160 km and 2,000 km above the Earth's surface.
*   **Characteristics:** Satellites in LEO have a shorter orbital period and orbit the Earth more frequently. They are used for various purposes, including Earth observation, remote sensing, and communication (e.g., Starlink and other satellite constellations).
*   **Presence in data:** Based on the bar chart, LEO is the most populated orbit in your dataset.

**Medium Earth Orbit (MEO):**
*   **Altitude:** Generally between 2,000 km and 35,786 km above the Earth's surface.
*   **Characteristics:** MEO satellites have a longer orbital period than LEO satellites. They are commonly used for navigation systems (like GPS, GLONASS, Galileo, and BeiDou) and some communication satellites.
*   **Presence in data:** MEO is the second most populated orbit in your dataset.

**Geostationary Orbit (GEO):**
*   **Altitude:** Approximately 35,786 km above the Earth's equator.
*   **Characteristics:** Satellites in GEO orbit the Earth at the same rate as the Earth rotates, appearing to stay in a fixed position relative to a point on the ground. This is ideal for communication satellites, especially for broadcasting, as ground antennas don't need to track the satellite.
*   **Presence in data:** GEO is the third most populated orbit in your dataset.

**Highly Elliptical Orbit (HEO):**
*   **Altitude:** Varies significantly throughout the orbit, with a high apogee (farthest point) and a low perigee (closest point).
*   **Characteristics:** HEOs are used for specific applications where a satellite needs to spend a long time over a particular region of the Earth, usually at high latitudes. This is common for some communication and scientific satellites.
*   **Presence in data:** HEO is the least populated orbit in your dataset.


Difference between the distinct satellites per year and the record per year
That's a great question! Here's the difference between "distinct satellites per year" and "records per year" based on how we calculated them:

Records per year: This counts every single row in your combined dataset for each year based on the 'Designator_year_launch' column. If a single satellite has multiple records in the data for a given year (perhaps from different epochs or updates), each of those records is counted. It tells you the total volume of data points available for each year.
Distinct satellites per year: This counts the number of unique satellite identifiers ('satNo') for each year based on the 'Designator_year_launch' column. If a satellite appears multiple times in the data for a given year, it is only counted once. This metric tells you how many different satellites are represented in the data for that year, regardless of how many records exist for each satellite.
In short, "records per year" is about the total number of data entries, while "distinct satellites per year" is about the number of individual satellites represented in the data for that year. The plot we generated shows both trends, and you can see how the number of records is generally higher than the number of distinct satellites because there are often multiple records for the same satellite within a year.

# Task
Analyze and visualize the age distribution of satellites and the trend of active vs. inactive satellites over time using the `df_combined_new` DataFrame.

## Calculate satellite age

### Subtask:
Create a new column in `df_combined_new` that represents the age of each satellite based on the difference between the current year and the 'Designator_year_launch'.


**Reasoning**:
Calculate the satellite age by subtracting the launch year from the current year and store it in a new column.



```python
import datetime

current_year = datetime.datetime.now().year
df_combined_new['satellite_age'] = current_year - df_combined_new['Designator_year_launch']
```

## Visualize age distribution

### Subtask:
Generate a histogram to show the distribution of satellite ages.


**Reasoning**:
Generate a histogram to visualize the distribution of satellite ages.



```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
df_combined_new['satellite_age'].hist(bins=50)
plt.xlabel('Satellite Age (Years)')
plt.ylabel('Number of Satellites')
plt.title('Distribution of Satellite Ages')
plt.grid(True)
plt.show()
```

## Identify active satellites

### Subtask:
Identify active satellites based on the 'epoch' column by considering satellites with 'epoch' values within the last two years as active.


**Reasoning**:
Convert the 'epoch' column to datetime objects and determine the date two years prior to the latest epoch to identify active satellites.



```python
df_combined_new['epoch'] = pd.to_datetime(df_combined_new['epoch'])
latest_epoch = df_combined_new['epoch'].max()
two_years_ago = latest_epoch - pd.DateOffset(years=2)
df_combined_new['is_active'] = df_combined_new['epoch'] > two_years_ago
```

## Analyze active vs. inactive over time

### Subtask:
Group the data by year and satellite status (active/inactive) and count the number of satellites in each category.


**Reasoning**:
Group the data by launch year and active status, count the occurrences, and reset the index for further processing.



```python
satellite_status_by_year = df_combined_new.groupby(['Designator_year_launch', 'is_active']).size().reset_index(name='satellite_count')
satellite_status_by_year.rename(columns={'Designator_year_launch': 'launch_year'}, inplace=True)
```

## Visualize active vs. inactive trends

### Subtask:
Plot the number of active and inactive satellites over time to show the trends.


**Reasoning**:
Generate a line plot to visualize the number of active and inactive satellites over time.



```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
for status in satellite_status_by_year['is_active'].unique():
    subset = satellite_status_by_year[satellite_status_by_year['is_active'] == status]
    label = 'Active' if status else 'Inactive'
    plt.plot(subset['launch_year'], subset['satellite_count'], marker='o', linestyle='-', label=label)

plt.title('Active vs. Inactive Satellite Trends Over Time')
plt.xlabel('Launch Year')
plt.ylabel('Number of Satellites')
plt.grid(True)
plt.legend()
plt.show()
```

## Summary:

### Data Analysis Key Findings

*   The age distribution of satellites shows a range of ages, with a histogram illustrating the frequency of satellites at different age intervals.
*   Satellites were categorized as active if their 'epoch' was within the last two years of the latest 'epoch' in the dataset, and inactive otherwise.
*   Analysis of satellite status over time reveals the count of active and inactive satellites launched each year.
*   A line plot visualizes the trends of active and inactive satellites based on their launch year, showing how the number of satellites in each status category has changed over time.

### Insights or Next Steps

*   Further analysis could explore the reasons behind satellites becoming inactive, such as reaching the end of their operational life or technical failures.
*   Investigating the distribution of satellite ages within specific orbital regimes or by satellite purpose could provide more detailed insights into the satellite population.

Analyze the density hotspots of satellites by binning altitude and inclination, creating a heatmap of satellite counts in each bin combination, and highlighting congested shells.

## Bin altitude and inclination

### Subtask:
Create new columns in `df_combined_new` to represent binned altitudes and inclinations.


**Reasoning**:
Create binned columns for altitude and inclination to prepare for density analysis.



```python
df_combined_new['altitude_bin'] = pd.cut(df_combined_new['Altitude_Avg'], bins=range(0, 40000, 550))
df_combined_new['inclination_bin'] = pd.cut(df_combined_new['inclination'], bins=range(0, 181, 5))
```

## Group by bins

### Subtask:
Group the DataFrame by the altitude and inclination bins and count the number of satellites in each combination.


**Reasoning**:
Group the data by the created altitude and inclination bins and count the number of satellites in each bin combination.



```python
density_by_bin = df_combined_new.groupby(['altitude_bin', 'inclination_bin']).size().reset_index(name='satellite_count')
```

## Prepare for heatmap

### Subtask:
Pivot the grouped data to create a matrix suitable for a heatmap, with altitude bins as rows, inclination bins as columns, and satellite counts as values.


**Reasoning**:
Pivot the grouped data to create a matrix suitable for a heatmap.



```python
heatmap_data = density_by_bin.pivot(index='altitude_bin', columns='inclination_bin', values='satellite_count')
```

## Generate heatmap

### Subtask:
Create a heatmap using the pivoted data to visualize the density of satellites across different altitude and inclination combinations.


**Reasoning**:
Create a heatmap using the pivoted data to visualize the density of satellites across different altitude and inclination combinations.



```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, cmap='viridis')
plt.title('Satellite Density Heatmap (Altitude vs. Inclination)')
plt.xlabel('Inclination Bin (deg)')
plt.ylabel('Altitude Bin (km)')
plt.show()
```
